# A program for linear regression model using ML

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

file_path = '/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv'
df = pd.read_csv(file_path)

X = df['YearsExperience']
y = df['Salary']

answer = linregress(X, y)
print(answer[0])
print(answer[1])
print(answer[2])
slope, intercept, r_value, p_value, std_err = linregress(X, y)

print("Slope (m):", slope)
print("Intercept (c):", intercept)

y_predict = slope * X + intercept

plt.scatter(df['YearsExperience'], df['Salary'], color='black', s=5)
plt.plot(df['YearsExperience'], y_predict, color='blue', linewidth=3)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Linear Regression using scipy.stats')
plt.show()
```

# A program for linear regression model using scikit-learn but no machine learning

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

file_path = '/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv'
df = pd.read_csv(file_path)

X = df['YearsExperience'].values.reshape(-1, 1)
y = df['Salary'].values.reshape(-1, 1)

model = LinearRegression()

model.fit(X, y)

m = model.coef_[0][0]
c = model.intercept_[0]

print("Final values of m and c:", m, c)

y_predict = model.predict(X)

plt.scatter(X, y, color='black', s=5)
plt.plot(X, y_predict, color='blue', linewidth=3)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Linear Regression using scikit-learn (without manual optimization)')
plt.show()
```

# A program without using scikit-learn or machine learning for linear regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = '/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv'
df = pd.read_csv(file_path)

X = df['YearsExperience'].values.reshape(-1, 1)
y = df['Salary'].values.reshape(-1, 1)

X = (X - np.mean(X)) / np.std(X)

m = 0
c = 0

L = 0.01
epochs = 1000

n = float(len(X))

for i in range(epochs):
    Y_pred = m * X + c
    D_m = (-2/n) * np.sum(X * (y - Y_pred))
    D_c = (-2/n) * np.sum(y - Y_pred)
    m = m - L * D_m
    c = c - L * D_c

print("Final values of m and c:", m, c)

y_predict = m * X + c

plt.scatter(X, y, color='black', s=5)
plt.plot(X, y_predict, color='blue', linewidth=3)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Linear Regression without scikit-learn or ML (using Gradient Descent)')
plt.show()
```
